```python
def human_readable_size(df):
    size_in_bytes = df.memory_usage(index=True).sum()
    for unit in ['B', 'KB', 'MB', 'GB', 'TB']:
        if size_in_bytes < 1024:
            return f"{size_in_bytes:.2f} {unit}"
        size_in_bytes /= 1024
human_readable_size(draft_chunk)
```

```python
cards = pd.read_csv("../data/cards.csv")[cards["expansion"] == "LTR"]

cards.head()
```

```python
# Space improvement from selecting type.
pool_chunk = draft_chunk[[col for col in draft_chunk.columns if col.startswith("pool_")]]
efficient_pool_chunk = pool_chunk.astype('Int8')
print(human_readable_size(pool_chunk), "->", human_readable_size(efficient_pool_chunk))

card_chunk = draft_chunk[[col for col in draft_chunk.columns if col.startswith("pack_card_")]]
efficient_card_chunk = pool_chunk.astype('bool')
print(human_readable_size(card_chunk), "->", human_readable_size(efficient_card_chunk))
```
