# Create Cube Cards File

Ad-hoc notebook to create CUBE cards.csv file

Needed because there is no CUBE set

```python
import pandas as pd

# Get list of all cards. 
df_cards = pd.read_csv("../data/cards/cards.csv")
df_cards.head()
```

```python
cube_picks = pd.read_csv("../data/17lands/draft_data_public.CUBE.PremierDraft.csv.gz", compression="gzip")
```

```python
card_names = [cn[5:] for cn in cube_picks.columns if cn.startswith("pool_")]

card_names = list(set(card_names))

for basic_land in ["Plains", "Island", "Swamp", "Mountain", "Forest"]:
    card_names.remove(basic_land)

len(card_names)
```

```python
cube_cards = df_cards[df_cards["name"].isin(card_names)].groupby("name").first().reset_index()

# Fix color identity
cube_cards["color_identity"] = cube_cards["color_identity"].fillna("Colorless")
cube_cards["color_identity"] = cube_cards["color_identity"].apply(
        lambda x: "Multicolor" if (len(x) > 1 and x != "Colorless") else x
    )

cube_cards = cube_cards.reset_index()[["name", "rarity", "color_identity"]]

cube_cards.head()
```

```python
cube_cards.to_csv("../data/cards/CUBE.csv", index=False)
```
